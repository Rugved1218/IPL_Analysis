```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

## Data Cleaning - Sayaji

print()


## Menu
def mainMenu():
    print("Ipl TOP 4 Teams Analysis")

    print()

    print("1. Player Stats ")
    print("2. Team Stats ")
    print("3. Rankings ")
    print("4. ROI")
    print("5. Exit")

    print()
    firstChoice=input("Enter your choice:")

    if(firstChoice.isdigit()):
        firstChoice=int(firstChoice)

    print()

    if(firstChoice==1):

        players=pd.read_csv(r"../datasets/iplAnalysisCleaned.csv")
        

        ## Player Stats - Vedant Agrawal

        print("Player Stats")

        print("1. Search Player ")
        print("2. Compare Player ")
        print("3. Filter by Role ")
        print("4. Back to main menu")
        print("5. Exit")

        print()
        
        secondChoice=input("Enter your choice:")

        if(secondChoice.isdigit()):
            secondChoice=int(secondChoice)

        print()

        if(secondChoice==1):
            ## Search Player
            print("Search Player")
            print()

            name=(input("Enter player name:").strip().title())
            
            # Function to show stats for any player
            def show_player_stats(name):
                player_stats = players[players["Player"] == name]
                if player_stats.empty:
                    print("Player not found.")
                else:
                    print(player_stats.transpose())
                
            show_player_stats(name);

        elif(secondChoice==2):
            ## Compare Player
            print("Compare Players")
            print()

            def comparePlayers():
                # Take player names input
                player1_name = (input("Enter the first player's name: ")).strip().title()
                player2_name = (input("Enter the second player's name: ")).strip().title()
                
                # Find the two players in the data
                player1_df = players[players['Player'] == player1_name]
                player2_df = players[players['Player'] == player2_name]
    
                # Check if the players exist
                if player1_df.empty:
                    print(f"First Player : {player1_name} not found")
                    return;
    
                if player2_df.empty:
                    print(f"Second Player : {player2_name} not found")
                    return;
    
                ## Get Player data
                player1=player1_df.iloc[0]
                player2=player2_df.iloc[0]
                
                # Print comparison of main stats
                print(f"\nComparison between {player1_name} and {player2_name}:")
                print(f"Runs scored: {player1['RunsScored']} vs {player2['RunsScored']}")
                print(f"Wickets taken: {player1['Wickets']} vs {player2['Wickets']}")
                print(f"Batting Average: {player1['BatAvg']} vs {player2['BatAvg']}")
                print(f"Bowling Average: {player1['BowlAvg']} vs {player2['BowlAvg']}")

            comparePlayers()

        

        elif(secondChoice==3):
            ## Filter By role

            # Menu loop
            while True:
                print("\nChoose a player role to show:")
                print("1 - Batsman")
                print("2 - Bowler")
                print("3 - All Rounder")
                print("4 - Wicket Keeper")
                print("5 - Exit")

                
                choice=input("Enter your choice:")

                if(choice.isdigit()):
                    choice=int(choice)
            
                if choice == 1:
                    role = "Batsman"
                elif choice == 2:
                    role = "Bowler"
                elif choice == 3:
                    role = "All Rounder"
                elif choice == 4:
                    role = "Wicket keeper"    
                elif choice == 5:
                    print("Exiting program.")
                    break
                else:
                    print("Invalid choice, please try again.")
                    continue

                # Filter players by role
                filtered_players = players[players['Role'] == role]
                
                # Display player names and basic stats
                print(f"\nList of {role}s:")
                for i in range(len(filtered_players)):
                    player = filtered_players.iloc[i]
                    print(f"{player['Player']} - Runs: {player['RunsScored']}, Wickets: {player['Wickets']}")

        
        elif(secondChoice==4):
                ## Back to main menu
                print("Main Menu")
                mainMenu()

        elif(secondChoice==5):
                ## Exit
                print("System Closed Succesfully")

        else:
                print("Invalid input: Enter the index as choice")

        ## End of Player Stats


    elif(firstChoice==2):

        ## Team Stats - Rugved Bajare


        print("Team Stats")

        ## Preparation

        #Teams
        players=pd.DataFrame()
        players=pd.read_csv(r"../datasets/iplAnalysisCleaned.csv")

        rcb=players[players["Team"]=="RCB"]
        pbks=players[players["Team"]=="PBKS"]
        gt=players[players["Team"]=="GT"]
        mi=players[players["Team"]=="MI"]

        #Functions
        def teamStat(team):
            teamName=team["Team"].iloc[0]
            
            totalRuns=int(team["RunsScored"].sum())
                
            totalWickets=int(team["Wickets"].sum())

            battingAvg=float(round((team["BatAvg"].sum())/(team["BatAvg"].count()),2))
                
            bowlEco=float(round((team["BowlEco"].sum())/(team["BowlEco"].count()),2))
                
            batSR=float(round((team["BatSR"].sum())/(team["BatSR"].count()),2))
                
            maxRuns=team["RunsScored"].max()
            bestBat=team[team["RunsScored"]==maxRuns]
            bestBatStr=bestBat["Player"].to_string(index=False)
                
            maxWickets=team["Wickets"].max()
            bestBowl=team[team["Wickets"]==maxWickets]
            bestBowlStr=bestBowl["Player"].to_string(index=False)

            teamData=[teamName,totalRuns,totalWickets,battingAvg,bowlEco,batSR,bestBatStr,bestBowlStr]
            return teamData

        def TeamSummary():
            TeamSummary=pd.DataFrame(data=[teamStat(rcb),teamStat(gt),teamStat(mi),teamStat(pbks)],columns=["Team","Total Runs","Total Wickets","Bat_Avg","Eco","Bat_Sr","Best Bat","Best Bowl"],index=[1,2,3,4])
            return TeamSummary
        
        def printingTeamStat(team):
        
            team=teamStat(team)
        
            teamName=team[0]
            print("Team Name :",teamName)

            totalRuns=team[1]
            print("Total Runs :",totalRuns)

            totalWickets=team[2]
            print("Total Wickets :",totalRuns)

            battingAvg=team[3]
            print("Batting Avg :",battingAvg)

            eco=team[4]
            print("Bowling Economy :",eco)

            batSR=team[5]
            print("Bat SR :",batSR)

            bestBat=team[6]
            print("Best Batsman :",bestBat)

            bestBowl=team[7]
            print("Best Bowler :",bestBowl)

            return 0

        def pieChartRoles(team):
            totalBatsman=team[team["Role"]=="Batsman"]["Role"].count();
            totalBowlers=team[team["Role"]=="Bowler"]["Role"].count();
            totalAllRounders=team[team["Role"]=="All Rounder"]["Role"].count();
            totalWicketKeepers=team[team["Role"]=="Wicket keeper"]["Role"].count();

            totalPlayersRole=[totalBatsman,totalBowlers,totalAllRounders,totalWicketKeepers] ##values
            colors=["red","blue","orange","green"];
            labels=["Batsman","Bowlers","All rounders","Wicket keepers"];
            plt.pie(totalPlayersRole,labels=labels,colors=colors,autopct='%1.1f%%')
            plt.title("Player Roles")
            plt.show()
            
            
        
        ## Menu

        print("1. Squads ")
        print("2. Team Summary ")
        print("3. Overall Team Summary ")
        print("4. Playoffs Summary")
        print("5. League Team Averages")
        print("6. Back to main menu")
        print("7. Exit")

        print()

        secondChoice=input("Enter your choice:")

        if(secondChoice.isdigit()):
            secondChoice=int(secondChoice)

        print()

        

        if(secondChoice==1):
            ## Squads
            print("Squads")

            print("1. RCB ")
            print("2. PBKS ")
            print("3. MI ")
            print("4. GT ")
            print("5. Back to main menu")
            print("6. Exit")

            print()

            thirdChoice=input("Enter your choice:")

            if(thirdChoice.isdigit()):
                thirdChoice=int(thirdChoice)

            print()

            if(thirdChoice==1):
                print("RCB Squad")
                print(rcb[["Player","Role"]])


            elif(thirdChoice==2):
                print("PBKS Squad")
                print(pbks[["Player","Role"]])
            
            elif(thirdChoice==3):
                print("MI Squad")
                print(mi[["Player","Role"]])
                
            
            elif(thirdChoice==4):
                print("GT Squad")
                print(gt[["Player","Role"]])

            elif(thirdChoice==5):
                ## Back to main menu
                print("Main Menu")
                mainMenu()
            
            elif(thirdChoice==6):
                ## Exit
                print("System Closed Succesfully")

            else:
                print("Invalid input: Enter the index as choice")


        elif(secondChoice==2):
            ## Team Summary

            print("Team Summary")

            print("1. RCB ")
            print("2. PBKS ")
            print("3. MI ")
            print("4. GT ")
            print("5. Back to main menu")
            print("6. Exit")

            print()

            thirdChoice=input("Enter your choice:")

            if(thirdChoice.isdigit()):
                thirdChoice=int(thirdChoice)

            print()

            if(thirdChoice==1):
                print("RCB Team Summary")
                printingTeamStat(rcb)
                print()
                pieChartRoles(rcb)


            elif(thirdChoice==2):
                print("PBKS Team Summary")
                printingTeamStat(pbks)
                print()
                pieChartRoles(pbks)

            
            elif(thirdChoice==3):
                print("MI Team Summary")
                printingTeamStat(mi)
                print()
                pieChartRoles(mi)

                
            elif(thirdChoice==4):
                print("GT Team Summary")
                printingTeamStat(gt)
                print()
                pieChartRoles(gt)

            elif(thirdChoice==5):
                ## Back to main menu
                print("Main Menu")
                mainMenu()
            
            elif(thirdChoice==6):
                ## Exit
                print("System Closed Succesfully")

            else:
                print("Invalid input: Enter the index as choice") 

                
        elif(secondChoice==3):
            ## Overall Team Summary
            print(TeamSummary())

        elif(secondChoice==4):
                ##League Summary

            def printLeagueSummary():
                print("League Summary")
                print()
                
                Summary=TeamSummary()

                totalRuns=int(Summary["Total Runs"].sum()) 
                print("Total Runs :",totalRuns) 

                totalWickets=int(Summary["Total Wickets"].sum()) 
                print("Total Wickets :",totalWickets) 

                total50s=players["50+ Scores"].sum()
                print("Total half centuries :",total50s)
                
                total100s=players["100+ Scores"].sum()
                print("Total Centuries :",total100s)

                total4W=players["4W"].sum()
                print("Total 4 Wicket Hauls :",total4W)

                total5W=players["5W"].sum()
                print("Total 5 Wicket Hauls :",total5W)

                print()

                max_runs=players["RunsScored"].max()
                bestBat=players[players["RunsScored"]==max_runs] 
                bestBatList=bestBat["Player"].to_list()
                bestBatStr=" , ".join(bestBatList)
                print(f"Highest Run Scorer : {bestBatStr} ({max_runs} runs) ")

                max_wickets=players["Wickets"].max()
                bestBowl=players[players["Wickets"]==max_wickets] 
                bestBowlList=bestBowl["Player"].to_list()
                bestBowlStr=" , ".join(bestBowlList)
                print(f"Highest Wicket Taker : {bestBowlStr} ({max_wickets} wickets)")

                max_50s=players["50+ Scores"].max()
                mostFiftiesPlayer=players[players["50+ Scores"]==max_50s] 
                mostFiftiesPlayerList=mostFiftiesPlayer["Player"].to_list()
                mostFiftiesPlayerStr=" , ".join(mostFiftiesPlayerList)
                print(f"Most Half Centuries : {mostFiftiesPlayerStr}({max_50s} Half Centuries)")

                max_100s=players["100+ Scores"].max()
                mostCenturiesPlayer=players[players["100+ Scores"]==max_100s] 
                mostCenturiesPlayerList=mostCenturiesPlayer["Player"].to_list()
                mostCenturiesPlayerStr=" , ".join(mostCenturiesPlayerList)
                print(f"Most Centuries  : {mostCenturiesPlayerStr} ({max_100s} Centuries)")
            

            printLeagueSummary()


        elif(secondChoice==5):
            ## League Team Averages

            def printLeagueAverage():
                Summary=TeamSummary()

                totalRuns=int(Summary["Total Runs"].mean())
                print("Total Runs :",totalRuns)

                totalWickets=int(Summary["Total Wickets"].mean())
                print("Total Runs :",totalWickets)

                battingAvg=round(Summary["Bat_Avg"].mean(),2)
                print("Batting Avg :",battingAvg)

                eco=round(Summary["Eco"].mean(),2)
                print("Bowling Economy :",eco)

                batSR=round(Summary["Bat_Sr"].mean(),2)
                print("Bat SR :",batSR)

            printLeagueAverage()
            print("League Team Average")
            print()
            pieChartRoles(players) ##for showing total player role percentages



        elif(secondChoice==6):
                ## Back to main menu
                print("Main Menu")
                mainMenu()
            
        elif(secondChoice==7):
                ## Exit
                print("System Closed Succesfully")

        else:
                print("Invalid input: Enter the index as choice") 

        print()


        
    elif(firstChoice==3):
        ## Rankings - Vard
        
        print("Rankings")
        print()

        players=pd.read_csv(r"../datasets/iplAnalysisCleaned.csv")

        # TOP 5 RUNSCORERS-
        toprunscorers = players.sort_values(by = 'RunsScored' , ascending=False)
        top5runscorers=toprunscorers.head(5)

        
        ##TOP WICKET TAKERS-
        topwickettakers = players.sort_values(by = 'Wickets' , ascending=False)
        top5wickettakers=topwickettakers.head(5)

       
        ##TOP 5 STRIKE HITTERS-
        topstrikehitters = players.sort_values(by = 'BatSR' , ascending=False)
        top5strikehitters=topstrikehitters.head(5)

        
        ## Worst 5 BOWLING ECONOMY-
        worstbowlingeco = players.sort_values(by = 'BowlEco' , ascending=False)
        worst5bowlingeco=worstbowlingeco.head(5)

        
        print("1. Top 5 Run Scorers")
        print("2. Top 5 Wicket Takers")
        print("3. Top 5 Strike Hitters")
        print("4. Worst 5 Bowling Eco")
        print("5. Back to main menu")
        print("6. Exit")
        print()

        
        secondChoice=input("Enter your choice:")

        if(secondChoice.isdigit()):
                secondChoice=int(secondChoice)

        print()
        
        if(secondChoice==1):
            print("Top 5 Run Scorers")
            print(top5runscorers[["Player","RunsScored"]].head(5))
            print()

            #Graph-
            
            runscorer=np.array(top5runscorers["Player"])
            runs=np.array(top5runscorers["RunsScored"])

            plt.bar(runscorer,runs)
            plt.title("Top 5 Run Scorers")
            plt.xlabel("Players")
            plt.ylabel("Runs Scored")
            plt.tight_layout()
            plt.xticks(rotation=45, ha='right')
            plt.show()

        
        elif(secondChoice==2):
            print("Top 5 Wicket Takers")
            print(topwickettakers[["Player","Wickets"]].head(5))
            print()

            #Graph-

            wickettaker=np.array(top5wickettakers["Player"])
            wickets=np.array(top5wickettakers["Wickets"])

            plt.bar(wickettaker,wickets)
            plt.title("Top 5 Wicket Takers")
            plt.xlabel("Players")
            plt.ylabel("Wickets Taken")
            plt.tight_layout()
            plt.xticks(rotation=45, ha='right')
            plt.show()

        
        elif(secondChoice==3):
            print("Top 5 Strike Hitters")
            print(topstrikehitters[["Player","BatSR"]].head(5))

            #Graph-
    
            strikehitter=np.array(top5strikehitters["Player"])
            strike_rate=np.array(top5strikehitters["BatSR"])

            plt.bar(strikehitter,strike_rate)
            plt.title("Top 5 Strike Hitters")
            plt.xlabel("Players")
            plt.ylabel("Strike Rate")
            plt.tight_layout()
            plt.xticks(rotation=45, ha='right')
            plt.show()

    
        elif(secondChoice==4):
            print("Worst 5 Bowling Eco")
            print(worst5bowlingeco[["Player","BowlEco"]])
            print()

            #Graph-

            bowlingeconomy=np.array(worst5bowlingeco["Player"])
            bowlers=np.array(worst5bowlingeco["BowlEco"])

            plt.bar(bowlingeconomy,bowlers)
            plt.title("Worst 5 Bowling Eco")
            plt.xlabel("Players")
            plt.ylabel("Bowling Economy")
            plt.tight_layout()
            plt.xticks(rotation=45, ha='right')
            plt.show()

        elif(secondChoice==5):
            print()
            print("Main Menu")
            mainMenu()
       
        elif(secondChoice==6):
            print("System Closed Successfully")

        else:
                print("Invalid input: Enter the index as choice")


    elif(firstChoice==4):
        ## ROI - Ritesh

        ## Preparation
        df=pd.read_csv(r"../datasets/iplAnalysisCleaned.csv")

        ## Adding Prices
        df.loc[df["Team"]=="RCB","Price"]=[21,11,2,11,3,11,1.50,5.75,10.75,5.00,12.50];
        df.loc[df["Team"]=="GT","Price"]=[8.50,16.50,15.75,2.60,4.00,4.00,18.00,1.30,2.00,9.50,12.25];
        df.loc[df["Team"]=="MI","Price"]=[1.00,16.30,8.00,16.35,16.35,5.25,1.00,2.00,9.25,12.50,18.00];
        df.loc[df["Team"]=="PBKS","Price"]=[4.00,3.80,2.60,26.75,4.20,5.50,2.40,7.00,1.50,18.00,18.00];

        ## Adding ROI
        import math

        Price=df["Price"]
        Runs=df["RunsScored"];
        Wickets=df["Wickets"];
        Matches=df["Matches"]
        StrikeRate=df["BatSR"];
        economy=df["BowlEco"]
        Fifties=df["50+ Scores"];
        Hundreds=df["100+ Scores"];
        fourfer=df["4W"];
        fifers=df["5W"];
        
        performance=( (Runs/Matches)*1.2 + (Wickets/Matches)*25 + (Fifties*5)+(Hundreds*10)+(fourfer*7)+(fifers*10))*2
        performanceValue=(Matches/(Matches+5))*performance
        baseline=(performanceValue.mean()/(Price.median() * 0.9))*0.5
        expectedValue=baseline*(Price**0.88)
        df["ROI"]=round(((performanceValue-expectedValue)/(expectedValue+50))*100,2)

        ##Converting to csv again for future use

        df.to_csv(r"../datasets/iplAnalysisPriceROI.csv")

        ## Menu
        
        print("ROI (Return on investment)")

        print("1. Auction Summary ")
        print("2. Player wise ROI table ")
        print("3. Top 5 ROI Players ")
        print("4. Worst 5 ROI Players")
        print("5. Back to main menu")
        print("6. Exit")

        print()

        players=pd.read_csv("../datasets/iplAnalysisPriceROI.csv")
        players_sorted = players.sort_values(by="ROI", ascending=False)
    
        secondChoice=input("Enter your choice:")

        if(secondChoice.isdigit()):
                secondChoice=int(secondChoice)

        print()
        
        if(secondChoice==1):
            print("Auction Summary")
            print()
            
            total_players = players["Player"].count()
            print(f"Total Players: {total_players} players")
            
            total_price = players["Price"].sum()
            print(f"Total Price spent in Auction: {total_price} crores")

            avg_price_per_player = round(total_price / total_players,2)
            print(f"Average Price spent in Auction per Player: {avg_price_per_player} crores")

            total_roi = players["ROI"].sum()
            avg_roi_per_player = round(total_roi / total_players,2)
            print(f"Average ROI per Player: {avg_roi_per_player} percent")
            
    
        elif(secondChoice==2):
            print("Player wise ROI table")
            print(players[["Player", "Team", "Price", "ROI"]])
           
        elif(secondChoice==3):
            
            print("TOP 5 BEST ROI Players:")

            ## Printing best ROI Players
            best5ROI=players_sorted[["Player", "Team", "Price", "ROI"]].head(5)
            print(best5ROI)
            print()

            ## Graph
            players = np.array(best5ROI["Player"])
            roi = np.array(best5ROI["ROI"])
                
            plt.bar(players,roi)
            plt.tight_layout()
            plt.xticks(rotation=45, ha='right')
            print()
            plt.title("TOP 5 BEST ROI Players")
            plt.xlabel("Players")
            plt.ylabel("ROI")
            plt.show()
            
        elif(secondChoice==4):

            ## Printing worst ROI Players
            print("TOP 5 WORST ROI Players:")
            worst5ROI=players_sorted[["Player", "Team", "Price", "ROI"]].tail(5)
            print(worst5ROI)
            print()

            ## Graph
            players = np.array(worst5ROI["Player"])
            roi = np.array(worst5ROI["ROI"])
            
            plt.bar(players,roi)
            plt.tight_layout()
            plt.xticks(rotation=45, ha='right')
            print()
            plt.title("TOP 5 WORST ROI Players")
            plt.xlabel("Players")
            plt.ylabel("ROI")
            plt.show()
                                                                                                                                                                                                                                    
        elif(secondChoice==5):
            print()
            print("Main Menu")
            mainMenu()
            
        elif(secondChoice==6):
            print("System Closed Successfully")

        else:
                print("Invalid input: Enter the index as choice")

    elif(firstChoice==5):
        ## Exit
        print("System Closed Succesfully")


    else:
        print("Invalid input: Enter the index as choice")

mainMenu()
```
